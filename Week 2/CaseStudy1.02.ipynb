```python
list2=list(healthc1['gender'])
genderr=[]
for i in list2:
    i == 'f' or i == 'female' or i == 'Female' or i== 'F':
        genderr.append('female')
    elif i == 'm'or i == 'male' or i == 'Male' or i== 'M':
        genderr.append('male')
    else:
        genderr.append('Unknown')
```

```python
gender_list = []
for i in range(len(healthc1["gender"])):
    if healthc1["gender"][i] == "male" or healthc1["gender"][i] == "Male" or healthc1["gender"][i] == "M": gender_list.append("male")
    elif healthc1["gender"][i] == "female" or healthc1["gender"][i] == "Female" or healthc1["gender"][i] == "F":
        gender_list.append("female")
    else:
        gender_list.append("unknown")
healthc1['gender'] = gender_list
print(healthc1['gender'])
```
